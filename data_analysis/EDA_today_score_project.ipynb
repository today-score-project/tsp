# 데이터 불러오기

```python
today = pd.read_excel("today_score_data.xlsx")
```

```python
quokka = today[today['name']=='quokka']
peacock =  today[today['name']=='peacock']
cheetah =  today[today['name']=='cheetah']
squirrel =  today[today['name']=='squirrel']
rabbit =  today[today['name']=='rabbit']
```

# EDA

```python
## 수면시간과 오늘의 점수
```

```python
f, axes = plt.subplots(2,3)

# 격자 크기 설정
f.set_size_inches((25,  13))

# 격자 여백 설정
plt.subplots_adjust(wspace=0.2, hspace=0.2)


## [0,0] 위치 quokka 수면시간&오늘의점수
axes[0,0].scatter(quokka['sleep_time'], quokka['today_score'], color = 'orange', alpha = 0.5, label = 'sleep_time')
axes[0,0].set_title('quokka 수면시간&오늘의점수', fontsize = 12)

## [0,1] 위치 peacock 수면시간&오늘의점수
axes[0,1].scatter(peacock['sleep_time'], peacock['today_score'], color = 'green', alpha = 0.5, label = 'sleep_time')
axes[0,1].set_title('peacock 수면시간&오늘의점수', fontsize = 12)

## [0,2] 위치 cheetah 수면시간&오늘의점수
axes[0,2].scatter(cheetah['sleep_time'], cheetah['today_score'], color = 'red', alpha = 0.5, label = 'sleep_time')
axes[0,2].set_title('cheetah 수면시간&오늘의점수', fontsize = 12)

## [1,0] 위치 squirrel 수면시간&오늘의점수
axes[1,0].scatter(squirrel['sleep_time'], squirrel['today_score'], color = 'skyblue', alpha = 0.5, label = 'sleep_time')
axes[1,0].set_title('squirrel 수면시간&오늘의점수', fontsize = 12)

## [1,1] 위치 rabbit 수면시간&오늘의점수
axes[1,1].scatter(rabbit['sleep_time'], rabbit['today_score'], color = 'gray', alpha = 0.7, label = 'sleep_time')
axes[1,1].set_title('rabbit 수면시간&오늘의점수', fontsize = 12)

##[1,2] 위치 모두 겹치기
axes[1,2].scatter(quokka['sleep_time'], quokka['today_score'], color = 'orange', alpha = 0.5, label = 'sleep_time', s=10)
axes[1,2].scatter(peacock['sleep_time'], peacock['today_score'], color = 'green', alpha = 0.5, label = 'sleep_time', s=10)
axes[1,2].scatter(cheetah['sleep_time'], cheetah['today_score'], color = 'red', alpha = 0.5, label = 'sleep_time', s=10)
axes[1,2].scatter(squirrel['sleep_time'], squirrel['today_score'], color = 'skyblue', alpha = 0.5, label = 'sleep_time',s=10 )
axes[1,2].scatter(rabbit['sleep_time'], rabbit['today_score'], color = 'gray', alpha = 0.5, label = 'sleep_time', s=10)
axes[1,2].set_title('모두 수면시간&오늘의점수', fontsize = 12)

plt.show()
```
